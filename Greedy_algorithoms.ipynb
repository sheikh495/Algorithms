```python
def visualize_change(n):
    # Define the denominations of coins and their names
    denominations = [(25, 'quarter'), (10, 'dime'), (5, 'nickel'), (1, 'penny')]

    # Initialize a dictionary to store the count of each coin used
    coin_counts = {}

    # Iterate through each denomination
    for denomination, coin_name in denominations:
        # Calculate the number of coins of the current denomination
        num_coins = n // denomination
        # Add the count of current denomination to the dictionary
        coin_counts[coin_name] = num_coins
        # Update the remaining amount after using the current denomination
        n -= num_coins * denomination

    # Print the visualization
    print("Visualization of change for", n, "cents:")
    for coin_name, count in coin_counts.items():
        print(count, coin_name, "coin(s)")

# Example usage
amount = 87  # Change for 87 cents
visualize_change(amount)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def print_tree(node, level=0, prefix="Root: "):
    if node is not None:
        print(" " * (level * 4) + prefix + str(node.value))
        print_tree(node.left, level + 1, "Left: ")
        print_tree(node.right, level + 1, "Right: ")

# Example usage
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("Visualizing the tree:")
print_tree(root)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def add_edges(node, graph, pos=None, x=0, y=0, layer=1):
    if pos is None:
        pos = {node.value: (x, y)}
    else:
        pos[node.value] = (x, y)

    if node.left is not None:
        graph.add_edge(node.value, node.left.value)
        lpos = (x - 1/layer, y - 1)
        pos = add_edges(node.left, graph, pos, lpos[0], lpos[1], layer + 1)

    if node.right is not None:
        graph.add_edge(node.value, node.right.value)
        rpos = (x + 1/layer, y - 1)
        pos = add_edges(node.right, graph, pos, rpos[0], rpos[1], layer + 1)

    return pos

def draw_tree(root):
    G = nx.DiGraph()
    pos = add_edges(root, G)

    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, with_labels=True, arrows=False, node_size=1000, node_color="lightblue", font_size=12)
    plt.show()

# Example usage
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("Graphical representation of the tree:")
draw_tree(root)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_change(n):
    # Define the denominations of coins and their names
    denominations = [(25, 'quarter'), (10, 'dime'), (5, 'nickel'), (1, 'penny')]

    # Create a directed graph
    G = nx.DiGraph()

    # Iterate through each denomination
    for denomination, coin_name in denominations:
        # Calculate the number of coins of the current denomination
        num_coins = n // denomination
        # Add nodes and edges to the graph
        G.add_node(coin_name)
        if num_coins > 0:
            G.add_edge(coin_name, 'Amount Left: {}'.format(n - num_coins * denomination), weight=num_coins)
        # Update the remaining amount after using the current denomination
        n -= num_coins * denomination

    # Plot the graph
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.title('Change Distribution Tree')
    plt.show()

# Example usage
amount = 87  # Change for 87 cents
visualize_change(amount)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_change(n):
    # Define the denominations of coins and their names
    denominations = [(25, 'quarter'), (10, 'dime'), (5, 'nickel'), (1, 'penny')]

    # Create a directed graph
    G = nx.DiGraph()

    # Iterate through each denomination
    for denomination, coin_name in denominations:
        # Calculate the number of coins of the current denomination
        num_coins = n // denomination
        # Add nodes and edges to the graph
        G.add_node(coin_name, label='{} ({})'.format(coin_name, num_coins))
        if num_coins > 0:
            G.add_edge(coin_name, 'Amount Left: {}'.format(n - num_coins * denomination), weight=num_coins)
        # Update the remaining amount after using the current denomination
        n -= num_coins * denomination

    # Plot the graph
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G)
    labels = nx.get_edge_attributes(G, 'weight')
    node_labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, labels=node_labels, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.title('Change Distribution Tree')
    plt.show()

# Example usage
amount = 87  # Change for 87 cents
visualize_change(amount)
```
